```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sn
import numpy as np
matplotlib.rc('font', size=16)
```

```python
df = pd.read_csv("saves/scores_queens.csv", sep=";")
```

```python
df["log10 convergence_time"] = np.log10(df["convergence_time (s)"])
df["log10 nodes open"] = np.log10(df["n_nodes_open"])

df=df.replace({"mode_val_heuristic": {1.0 :"First value", 2.0 :"Most supported value"}})

df=df.replace({"mode_var_heuristic": {1.0 : "First variable", 2.0 :"Smallest domain variable", 3.0 : "Most constrained variable", 4.0 : "Most linked to the instanciated variables"}})
```

```python
df.head()
```

```python
plt.rc('legend', fontsize=12)    # legend fontsize
fig, ax = plt.subplots(1, 1, figsize=(15, 8))
sn.lineplot(data=df, x='nqueens', y="log10 convergence_time", hue=df[['mode_var_heuristic', 'mode_val_heuristic']].apply(tuple, axis=1), markers=True, ax = ax, alpha=1)
ax.set_ylabel(f" Log10 of the Execution Time (s)")
ax.set_xlabel("Number of Queens")
ax.grid(True)
ax.set_title("Log10 of the execution time according to the number of queens\n for different settings of the solver")
```

```python
plt.rc('legend', fontsize=12)    # legend fontsize
fig, ax = plt.subplots(1, 1, figsize=(15, 8))
sn.lineplot(data=df, x='nqueens', y="log10 nodes open", hue=df[['mode_var_heuristic', 'mode_val_heuristic']].apply(tuple, axis=1), markers=True, ax = ax, alpha=1)
ax.set_ylabel(f"Log10 of the Number of nodes open")
ax.set_xlabel("Number of Queens")
ax.grid(True)
```
